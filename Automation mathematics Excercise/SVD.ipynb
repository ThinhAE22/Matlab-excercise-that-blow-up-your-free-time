```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
#%%
X,y=datasets.fetch_olivetti_faces(return_X_y=True)

#X=400x4096-matrix, datapoints as rows
#y=0,1,2,...,39=person number

kuvak =[]
plt.figure(figsize=(10, 10))
for k in range(0, 399, 10):
    kuvak.append(X[k, :].reshape(64, 64))

for i in range(40):
    plt.subplot(5, 8, i+1)
    plt.imshow(kuvak[i], cmap='gray')
    plt.title('y = {}'.format(i))
    plt.axis('off')
```

```python
# Perform Singular Value Decomposition (SVD)
mu=np.mean(X,axis=0)
Xhat=X-mu
U, s, V = np.linalg.svd(Xhat, full_matrices=False)

# Plot the singular values
plt.figure(figsize=(8,4))
plt.plot(s)
plt.grid()
plt.xlabel('k', fontsize=14)
plt.ylabel('Singular Value $\sigma_k$', fontsize=14)
plt.show()
```

```python
vpcx = np.cumsum(s**2)
vpcy = np.sum(s**2)
plt.figure(figsize=(8,4))
plt.plot(vpcx/vpcy)
plt.grid()
plt.xlabel('k', fontsize=14)
plt.title('varience of the principal component 1...k', fontsize=14)
plt.show()
```
